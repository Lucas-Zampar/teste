```python
import ipywidgets as widgets 
```

```python
widget_massa = widgets.FloatText(description="m: ") #widget para captuar massa
widget_gravidade = widgets.FloatText(description="g: ", value=10) #widget para captuar gravidade
widget_label = widgets.Label(value="Peso: ") #widget para exibir o peso
```

```python
widget_botao = widgets.Button(description="Calcular Peso") #widget do botão
```

```python
#define a função que calculará o peso
def calcular_peso(massa, gravidade):
  return massa * gravidade
```

```python
#define a função executada no clique do botão
def ao_clicar(b):
  massa = widget_massa.value
  gravidade = widget_gravidade.value

  peso = calcular_peso(massa, gravidade)

  rotulo = f"Peso: {peso:.2f}"
  widget_label.value = rotulo
```

```python
#relaciona a função acima com o clique do botão
widget_botao.on_click(ao_clicar)
```

```python
#declara uma lista de widgets ordenados conforme exibição
lista_widgets = [widget_massa, widget_gravidade, widget_label, widget_botao]
```

```python
#a lista é repassada para um VBox
vbox = widgets.VBox(lista_widgets)
```

```python
#a interface final é exibida
display(vbox)
```
